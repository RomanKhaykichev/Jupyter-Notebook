- 1.1 Изучите количество памяти с помощью matplotlib


```python
def memory(memory_str):
    gb = memory_str.split(' ')[0]
    if gb.endswith('GB'):
        gb = float(gb[:-2])
    elif gb.endswith('TB'):
        gb = float(gb[:-2])*1024
    return gb


df['Memory_Amount'] = df['Memory'].apply(memory)
df.head(3)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Memory_Amount'])
plt.title("Memory")
plt.xlabel('GB')
plt.ylabel('Count')
plt.xticks(rotation=30);
```

### Задание 3.
Изучите взаимосвязь компаний производителей ноутбуков и
компаний производителей процессоров, используя сложенную
или многорядовую столбчатую диаграмму
Процессоры от Samsung не изучайте

3.1 Постройте график в абсолютных величинах

```python
data_5 = pd.crosstab(index=df['Company'], columns=df['Cpu_Company'])
data_5 = data_5.drop(columns=['Samsung']).reset_index()
data_5
```

```python
# Строим столбчатую многорядную диаграмму
# определяем кол-во делений
n_ticks = np.arange(len(data_5['AMD']))
# определяем сдвиг
offset = 0.2
# определяем ширину столбцов
w = 0.4
# добавляем сдвиг к кол-ву делений
plt.bar(n_ticks-offset, data_5['AMD'], width=w)
plt.bar(n_ticks+offset, data_5['Intel'], width=w)
plt.title('Notebook')
plt.xlabel('Company')
plt.ylabel('Count')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, data_5['Company'], rotation=60);  # добавляем метки делений
```

3.2 Постройте график в относительных величинах

```python
# Строим столбчатую многорядную сложенную диаграмму
plt.figure(figsize=(8, 6))

plt.bar(n_ticks, data_5['AMD'])
plt.bar(n_ticks, data_5['Intel'], bottom=data_5['AMD'])

plt.title('Notebook')
plt.xlabel('Company')
plt.ylabel('Count')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, data_5['Company'], rotation=60);
```

### Задание 5*.
Постройте матрицу корреляций для таблицы

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.5] = 0
corr_matrix
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title("Correlation matrix")
plt.xticks(rotation=60);
```
